```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv("C:/Users/LENOVO/Downloads/Student Evaluation.csv")
```

Task 1: Data Cleaning

```python
df.head() #shows first 5 rows.
```

```python
df.isnull().sum() #Identify missing values in the dataset.
```

```python
mean_score = df['Score'].mean()  #find mean of score
mean_score
```

```python
df['Score']=df['Score'].fillna(mean_score)  #Replace missing "Score" values with the average score of that subject.
df
```

```python
mean_Hours_Studied = df['Hours_Studied'].mean() #find mean of hours studied
mean_Hours_Studied
```

```python
df['Hours_Studied']=df['Hours_Studied'].fillna(mean_Hours_Studied) #Replace missing "Hours_Studied" values with the mean of "Hours_Studied".
df
```

```python
df.dropna(subset=['Score', 'Hours_Studied'], how='all', inplace=True) #Drop rows where both "Score" and "Hours_Studied" are missing (if any).
df
```

Task 3: Subject-wise Analysis

```python
sub_mean = df.groupby("Subject")["Score"].mean() #Calculate the average "Score" for each subject.
```

```python
sub_mean
```

```python
highest_score_sub = sub_mean.idxmax() #Which subject has the highest average "Score"?
highest_score = sub_mean.max()
print(f"Subject which has highest score is {highest_score_sub} with {highest_score} score")
```

```python
hrs_mean = df.groupby("Subject")["Hours_Studied"].mean()
```

```python
hrs_mean
```

```python
highest_hrs_sub = hrs_mean.idxmax() #Find the subject where students studied the most hours on average.
highest_hrs = hrs_mean.max()
print(f"Subject where students studied the most is {highest_hrs_sub} with {highest_hrs} hrs of study")
```

```python
top = df.groupby("Subject")["Score"].idxmax() #Identify the top performer (highest score) in each subject.
```

```python
top
```

```python
top_performer = df.loc[top, ["Subject", "Name", "Score"]]
```

```python
top_performer
```

Task 4: Regional Analysis

```python
region_mean = df.groupby("Region")["Score"].mean()  #Find the average "Score" of students in each region.
region_mean 
```

```python
avg_hrs = df.groupby("Region")["Hours_Studied"].mean() #Which region has the highest average study hours?
avg_hrs
```

```python
highest_avg_hrs = avg_hrs.max()
highest_hrs_rg = avg_hrs.idxmax()
print(f"Highest average studied region is {highest_hrs_rg} with {highest_avg_hrs} of hours")
```

```python
rg_mean = df.groupby("Region")["Score"].mean() # Identify the region with the lowest overall performance.
rg_mean
```

```python
lowest_rg = rg_mean.idxmin()
lowest_rg_mean = rg_mean.min()
print(f"Lowest performance region is {lowest_rg} with {lowest_rg_mean} scores")
```

Task 5: Gender-Based Analysis

```python
gender_mean = df.groupby("Gender")["Score"].mean() #Compare the average "Score" of male vs female students.
gender_mean
```

```python
avg_gen_hrs = df.groupby("Gender")["Hours_Studied"].mean() #Find which gender studied more hours on average.
high_gen = avg_gen_hrs.idxmax()
high_gen_mean = avg_gen_hrs.max()
print(f"{high_gen} has more average studied hours, {high_gen_mean} hours ")
```

```python
top3_female = df[df["Gender"] == "F"].nlargest(3, "Score") #Identify the top 3 highest-scoring female students.
top3_female
```

```python
lowest_2M = df[df["Gender"] == "M"].nsmallest(2, "Score") #Find the lowest 2 scoring male students.
lowest_2M
```

Task 6: Performance Categorization

```python
df["Performance"] = df["Score"].apply(lambda x: "High" if x >= 80 else "Low")
```

```python
df
```

```python
performance_count = df["Performance"].value_counts() #Count how many students fall into "High" and "Low" categories.
performance_count
```

```python
performance_distribution = df.groupby(["Subject", "Performance"]).size().unstack(fill_value=0) #Compare the distribution of "Performance" across different subjects.
print(performance_distribution)
```

```python
import matplotlib.pyplot as plt

performance_distribution.plot(kind='bar', figsize=(8,5))
plt.title("Performance Distribution by Subject")
plt.ylabel("Number of Students")
plt.xlabel("Subject")
plt.show()
```
